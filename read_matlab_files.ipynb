# Simple data exploration

This notebook shows the structure of the data.

```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import pandas as pd
```

## Review `subject*.mat` files

```python
mat = loadmat('data_emg/subject1.mat')  # load mat-file
mat.keys()
```

```python
df = pd.DataFrame(mat["Mtrain"][0, :-1], columns=['Values'])
df

# Create a plot using pandas
df.plot(kind='line', figsize=(10, 6), grid=True, title='Example one time series')
plt.ylabel('Value')
plt.xlabel('Time')
plt.show()
```

```python
print("shape training:", mat["Mtrain"].shape)
print("shape validation:", mat["Mcv"].shape)
print("shape test:", mat["Mtest"].shape)
print("shape test:", [(f"label {idx + 1}", f"label name: {list(name)[0]}") for idx, name in enumerate(mat["label_names"][0])])
```
